```kotlin
/**
 * Обучение методом "холодно/горячо"
 */
var weight = 0.5
val input = 0.5
val goalPrediction = 0.5

val stepAmount = 0.001

for (iteration in 0..1101) {
    val prediction = input * weight
    val error = pow(prediction - goalPrediction, 2.0)

    println("Error: $error Predicton: $prediction")

    val upPrediction = input * (weight + stepAmount)
    val upError = pow(upPrediction - goalPrediction, 2.0)

    val downPrediction = input * (weight + stepAmount)
    val downError = pow(downPrediction - goalPrediction, 2.0)

    weight = if (downError < upError) {
        weight - stepAmount
    } else {
        weight + stepAmount
    }
}
```
